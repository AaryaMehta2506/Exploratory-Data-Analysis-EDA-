# Exploratory Data Analysis (EDA)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Visualization style

```python
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

### Load Dataset

```python
df = pd.read_csv("World Happiness Report.csv")
print(f"Dataset Shape: {df.shape}")
df.head()
```

### Inspect Dataset

```python
df.info()
df.describe()
```

### Missing Values Check

```python
print("Missing values per column:")
print(df.isnull().sum())

df = df.drop_duplicates()

# fill or drop missing values, e.g.:
# df['Happiness Score'] = df['Happiness Score'].fillna(df['Happiness Score'].median())
```

```python
print(df.columns)
```

### Univariate Analysis

```python
# column names based on your dataset
score_col = 'Happiness_Score'

# Distribution of Happiness Score
sns.histplot(df[score_col], kde=True)
plt.title('Distribution of Happiness Score')
plt.show()

# Top 10 Happiest Countries per Year
for year in sorted(df['Year'].unique()):
    top10 = df[df['Year'] == year].nlargest(10, score_col)
    sns.barplot(data=top10, x=score_col, y='Country', palette='viridis')
    plt.title(f'Top 10 Happiest Countries  {year}')
    plt.show()
```

### Bivariate Analysis

```python
numeric_cols = [
    'Happiness_Score', 'GDP_per_Capita', 'Social_Support', 
    'Health', 'Freedom', 'Trust', 'Generosity', 'Dystopia Residual'
]

# Correlation heatmap
corr = df[numeric_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# GDP vs Happiness
sns.scatterplot(data=df, x='GDP_per_Capita', y=score_col, hue='Year', palette='coolwarm')
plt.title('GDP per Capita vs Happiness Score')
plt.show()

# Health vs Happiness
sns.scatterplot(data=df, x='Health', y=score_col, hue='Year', palette='mako')
plt.title('Health (Life Expectancy) vs Happiness Score')
plt.show()
```

### Multivariate Insights

```python
# Pairplot: relationships among multiple numeric variables
sns.pairplot(df[numeric_cols], diag_kind='kde')
plt.suptitle('Pairplot of Key Numerical Features', y=1.02)
plt.show()

# 3D Scatterplot: GDP, Health, Happiness
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['GDP_per_Capita'], df['Health'], df[score_col],
           c=df[score_col], cmap='viridis', s=50)
ax.set_xlabel('GDP per Capita')
ax.set_ylabel('Health (Life Expectancy)')
ax.set_zlabel('Happiness Score')
plt.title('3D View: GDP, Health, and Happiness')
plt.show()

# FacetGrid: GDP vs Happiness by Region
g = sns.FacetGrid(df, col="Region", col_wrap=3, height=4, sharex=False, sharey=False)
g.map_dataframe(sns.scatterplot, x="GDP_per_Capita", y=score_col, hue="Year", palette="coolwarm")
g.add_legend()
plt.show()
```

### Regional Analysis

```python
avg_by_region = df.groupby('Region')[score_col].mean().sort_values(ascending=False)
sns.barplot(x=avg_by_region.values, y=avg_by_region.index, palette='plasma')
plt.title('Average Happiness Score by Region')
plt.show()
```

### Time Trends

```python
avg_by_year = df.groupby('Year')[score_col].mean().reset_index()
sns.lineplot(data=avg_by_year, x='Year', y=score_col, marker='o')
plt.title('Average Happiness Score Over Years')
plt.show()
```

### Automatic Insights Generator

```python
def generate_insights(df):
    insights = []
    
    # Correlation analysis
    corr = df[numeric_cols].corr()[score_col].sort_values(ascending=False)
    top_pos = corr.drop(score_col).head(3)
    bottom = corr.drop(score_col).tail(3)
    
    insights.append(" **Top Positive Correlations with Happiness Score:**")
    for feat, val in top_pos.items():
        insights.append(f"- {feat}: {val:.2f}")

    insights.append("\n **Lowest (or Negative) Correlations:**")
    for feat, val in bottom.items():
        insights.append(f"- {feat}: {val:.2f}")

    # Time trend
    trend = avg_by_year[score_col]
    if trend.is_monotonic_increasing:
        insights.append("\n Average happiness has increased over the years.")
    elif trend.is_monotonic_decreasing:
        insights.append("\n Average happiness has decreased over the years.")
    else:
        insights.append("\n Average happiness has fluctuated over the years.")

    return "\n".join(insights)

print(generate_insights(df))
```
